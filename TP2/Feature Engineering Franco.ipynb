````python
import datetime as datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
````

````python
df = pd.read_csv('dataset_v1')
df.info()
````

# Structures (categoricos)

````python
structures = df.loc[:, 'has_superstructure_adobe_mud' : 'has_superstructure_other']
struc = structures.head()
struc
````

````python
s = df.loc[:, 'has_superstructure_adobe_mud' : 'has_superstructure_stone_flag'].head()
s

col_mean = s['has_superstructure_adobe_mud'].mean()
s.loc[s['has_superstructure_adobe_mud'], 'has_superstructure_adobe_mud_mean'] = col_mean
s.loc[~s['has_superstructure_adobe_mud'], 'has_superstructure_adobe_mud_mean'] = 1 - col_mean
s
````

````python
cols = [
	'has_superstructure_adobe_mud',
	'has_superstructure_mud_mortar_stone',
	'has_superstructure_stone_flag',
	'has_superstructure_cement_mortar_stone',
	'has_superstructure_mud_mortar_brick',
	'has_superstructure_cement_mortar_brick',
	'has_superstructure_timber',
	'has_superstructure_bamboo',
	'has_superstructure_rc_non_engineered',
	'has_superstructure_rc_engineered',
	'has_superstructure_other'
]

for col_name in cols:
    col_mean = structures[col_name].mean()
    structures.loc[structures[col_name], col_name + '_mean'] = col_mean
    structures.loc[~structures[col_name], col_name + '_mean'] = 1 - col_mean


structures_slice = structures.loc[:, 'has_superstructure_adobe_mud_mean' : 'has_superstructure_other_mean']
structures_slice.info()
````

# Building categoricals

````python

cc = [ # cc: "categorical columns"
	'land_surface_condition',
	'foundation_type',
	'roof_type',
	'ground_floor_type',
	'other_floor_type',
	'position',
	'plan_configuration',
    'legal_ownership_status'
]

build_cats = df.loc[:, cc]

# land_surface_condition
build_cats[cc[0]].replace(
    {"n": cc[0]+" N", "o":  cc[0]+" O", "t": cc[0]+" T"},
    inplace=True
)
# foundation_type
build_cats[cc[1]].replace(
    {"h": cc[1]+" H", "i":  cc[1]+" I", "r": cc[1]+" R", "u": cc[1]+" U", "w": cc[1]+" W"},
    inplace=True
)
# roof_type
build_cats[cc[2]].replace(
    {"n": cc[2]+" N", "q":  cc[2]+" Q", "x": cc[2]+" X"},
    inplace=True
)
# ground_floor_type
build_cats[cc[3]].replace(
    {"f": cc[3]+" F", "m":  cc[3]+" M", "v": cc[3]+" V", "x": cc[3]+" X", "z": cc[3]+" Z"},
    inplace=True
)
# other_floor_type
build_cats[cc[4]].replace(
    {"j": cc[4]+" J", "q":  cc[4]+" Q", "s": cc[4]+" S", "x": cc[4]+" X"},
    inplace=True
)
# position
build_cats[cc[5]].replace(
    {"j": cc[5]+" H", "o":  cc[5]+" O", "s": cc[5]+" S", "t": cc[5]+" T"},
    inplace=True
)
# plan_configuration
build_cats[cc[6]].replace(
    {"a": cc[6]+" A", "c": cc[6]+" C", "d": cc[6]+" D", "f": cc[6]+" F", "m": cc[6]+" M",
     "n": cc[6]+" N", "o": cc[6]+" O", "q": cc[6]+" Q", "s": cc[6]+" S", "u": cc[6]+" U"},
    inplace=True
)
# legal_ownership_status
build_cats[cc[7]].replace(
    {"a": cc[7]+" A", "r":  cc[7]+" R", "v": cc[7]+" V", "w": cc[7]+" W"},
    inplace=True
)

for col_name in cc:
    dummies = pd.get_dummies(build_cats[col_name])
    build_cats = pd.concat([build_cats, dummies], axis=1)
    
build_cats = build_cats.drop(columns=cc)
build_cats.loc[build_cats['foundation_type H'] == 1]
````

Problema principal de hacer one hot encoding de esta forma: practicamente se duplica la cantidad de columnas

# Numericos

````python
nums = df.loc[:, ['area_percentage', 'height_percentage']]
nums.describe()
````

Podemos asumir que `area_percentage` es un porcentaje (el minimo es 1 y el maximo es 100), pero que pasa con `height_percentage`?

De todas formas podemos normalizar ambos a un numero entre 0 y 1 con:

```
norm(x) = (x - min) / (max - min)
```

````python
from sklearn import preprocessing

values = nums.values
scaler = preprocessing.MinMaxScaler()
scaled_values = scaler.fit_transform(values)
scaled_nums = pd.DataFrame(scaled_values)
scaled_nums.rename(
    columns={0: 'area_percentage_norm', 1: 'height_percentage_norm'},
    inplace=True
)
scaled_nums.head()
````

````python
df3 = df2.join(structures_slice)
df3 = df3.join(scaled_nums)
df3 = df3.drop(columns=[
    'Unnamed: 0', 
    'area_percentage',
    'height_percentage',
    'has_superstructure_adobe_mud',
    'has_superstructure_mud_mortar_stone',
    'has_superstructure_stone_flag',
    'has_superstructure_cement_mortar_stone',
    'has_superstructure_mud_mortar_brick',
    'has_superstructure_cement_mortar_brick',
    'has_superstructure_timber',
    'has_superstructure_bamboo',
    'has_superstructure_rc_non_engineered',
    'has_superstructure_rc_engineered',
    'has_superstructure_other',
    'has_secondary_use',
    
])
df3.info()
````
