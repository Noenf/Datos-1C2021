```python
df['roof_type_C1'] = 0
df['roof_type_C0'] = 0
df['roof_type_C1'] = df['roof_type_C1'].astype('int64') 
df['roof_type_C0'] = df['roof_type_C0'].astype('int64')
df.head()
```

```python
#land_surface_condition                  260601 non-null  object
#foundation_type                         260601 non-null  object
#roof_type                               260601 non-null  object
#ground_floor_type                       260601 non-null  object
#other_floor_type                        260601 non-null  object
#position                                260601 non-null  object
#plan_configuration                      260601 non-null  object
#legal_ownership_status                  260601 non-null  object

```

```python
#roof_type 
# valor n= 01
# valor q= 10
# valor x= 11
def binarioRoofType_C0(valor):
    if ((valor == 'n') or (valor == 'x')):
        return 1;
    else: return 0;
    
def binarioRoofType_C1(valor):
    if ((valor == 'q') or (valor == 'x')):
        return 1;
    else: return 0;
        
```

```python
df['roof_type_C1'] = df['roof_type'].apply(binarioRoofType_C1)
df['roof_type_C0'] = df['roof_type'].apply(binarioRoofType_C0)
df.head()
```

```python
df['land_surface_condition'].value_counts()
```

```python
#land_surface_condition 
# valor t= 01
# valor n= 10
# valor o= 11
def binario_land_surface_condition_C0(valor):
    if ((valor == 't') or (valor == 'o')):
        return 1;
    else: return 0;
    
def binario_land_surface_condition_C1(valor):
    if ((valor == 'n') or (valor == 'o')):
        return 1;
    else: return 0;


```

```python
df['land_surface_condition_C0'] =0
df['land_surface_condition_C1'] =0
```

```python
df['land_surface_condition_C1'] = df['land_surface_condition'].apply(binario_land_surface_condition_C1)
df['land_surface_condition_C0'] = df['land_surface_condition'].apply(binario_land_surface_condition_C0)
df.head()
```

```python
df['foundation_type'].value_counts()
```

```python

#foundation_type 
# valor r= 001
# valor w= 010
# valor u= 011
# valor i= 100
# valor n= 101

def binario_foundation_type_C0(valor):
    if ((valor == 'r') or (valor == 'u') or (valor == 'n')):
        return 1;
    else: return 0;
    
def binario_foundation_type_C1(valor):
    if ((valor == 'w') or (valor == 'u')):
        return 1;
    else: return 0;

def binario_foundation_type_C2(valor):
    if ((valor == 'i') or (valor == 'n')):
        return 1;
    else: return 0;
    

df['foundation_type_C0'] =0
df['foundation_type_C1'] =0
df['foundation_type_C2'] =0

df['foundation_type_C0'] = df['foundation_type'].apply(binario_foundation_type_C0)
df['foundation_type_C1'] = df['foundation_type'].apply(binario_foundation_type_C1)
df['foundation_type_C2'] = df['foundation_type'].apply(binario_foundation_type_C2)
df.head()
```

```python
df['ground_floor_type'].value_counts()
```

```python
#ground_floor_type
# valor f= 001
# valor x= 010
# valor v= 011
# valor z= 100
# valor m= 101

def binario_ground_floor_type_C0(valor):
    if ((valor == 'f') or (valor == 'v') or (valor == 'm')):
        return 1;
    else: return 0;
    
def binario_ground_floor_type_C1(valor):
    if ((valor == 'x') or (valor == 'v')):
        return 1;
    else: return 0;

def binario_ground_floor_type_C2(valor):
    if ((valor == 'z') or (valor == 'm')):
        return 1;
    else: return 0;
    

df['ground_floor_type_C0'] =0
df['ground_floor_type_C1'] =0
df['ground_floor_type_C2'] =0

df['ground_floor_type_C0'] = df['ground_floor_type'].apply(binario_ground_floor_type_C0)
df['ground_floor_type_C1'] = df['ground_floor_type'].apply(binario_ground_floor_type_C1)
df['ground_floor_type_C2'] = df['ground_floor_type'].apply(binario_ground_floor_type_C2)
df.head()
```

```python
df["other_floor_type"].value_counts()
```

```python
#other_floor_type
# valor s= 001
# valor j= 010
# valor x= 011
# valor q= 100

def binario_other_floor_type_C0(valor):
    if ((valor == 's') or (valor == 'x')):
        return 1;
    else: return 0;
    
def binario_other_floor_type_C1(valor):
    if ((valor == 'j') or (valor == 'x')):
        return 1;
    else: return 0;

def binario_other_floor_type_C2(valor):
    if ((valor == 'q')):
        return 1;
    else: return 0;
    

df['other_floor_type_C0'] =0
df['other_floor_type_C1'] =0
df['other_floor_type_C2'] =0

df['other_floor_type_C0'] = df['other_floor_type'].apply(binario_other_floor_type_C0)
df['other_floor_type_C1'] = df['other_floor_type'].apply(binario_other_floor_type_C1)
df['other_floor_type_C2'] = df['other_floor_type'].apply(binario_other_floor_type_C2)
df.head()
```

```python
#position
# valor t= 001
# valor s= 010
# valor j= 011
# valor o= 100

def binario_position_C0(valor):
    if ((valor == 't') or (valor == 'j')):
        return 1;
    else: return 0;
    
def binario_position_C1(valor):
    if ((valor == 's') or (valor == 'j')):
        return 1;
    else: return 0;

def binario_position_C2(valor):
    if ((valor == 'o')):
        return 1;
    else: return 0;
    

df['position_C0'] =0
df['position_C1'] =0
df['position_C2'] =0

df['position_C0'] = df['position'].apply(binario_position_C0)
df['position_C1'] = df['position'].apply(binario_position_C1)
df['position_C2'] = df['position'].apply(binario_position_C2)
df.head()
```

```python
df["plan_configuration"].value_counts()
```

```python
#plan_configuration
# valor d= 0001
# valor q= 0010
# valor u= 0011
# valor s= 0100
# valor c= 0101
# valor a= 0110
# valor o= 0111
# valor m= 1000
# valor n= 1001
# valor f= 1010

def binario_plan_configuration_C0(valor):
    if ((valor == 'd') or (valor == 'u') or (valor == 'c') or (valor == 'o') or (valor == 'n')):
        return 1;
    else: return 0;
    
def binario_plan_configuration_C1(valor):
    if ((valor == 'q') or (valor == 'u') or (valor == 'a') or (valor == 'o') or (valor == 'f')):
        return 1;
    else: return 0;

def binario_plan_configuration_C2(valor):
    if ((valor == 's') or (valor == 'c') or (valor == 'a') or (valor == 'o')):
        return 1;
    else: return 0;
    
def binario_plan_configuration_C3(valor):
    if ((valor == 'm') or (valor == 'n') or (valor == 'f')):
        return 1;
    else: return 0;
    

df['plan_configuration_C0'] =0
df['plan_configuration_C1'] =0
df['plan_configuration_C2'] =0
df['plan_configuration_C3'] =0

df['plan_configuration_C0'] = df['plan_configuration'].apply(binario_plan_configuration_C0)
df['plan_configuration_C1'] = df['plan_configuration'].apply(binario_plan_configuration_C1)
df['plan_configuration_C2'] = df['plan_configuration'].apply(binario_plan_configuration_C2)
df['plan_configuration_C3'] = df['plan_configuration'].apply(binario_plan_configuration_C3)

df.head()
```
