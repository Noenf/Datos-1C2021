```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('dataset_v4.csv')
```

```python
train = train.drop(labels=['building_id','geo_level_2_id','geo_level_3_id'],axis=1)
```

```python
train.info()
```

```python
test=pd.read_csv('encoded_test_values2.csv')
```

```python
test = test.drop(labels=['land_surface_condition','foundation_type','roof_type',\
'ground_floor_type','other_floor_type','position', 'plan_configuration','legal_ownership_status',\
 'has_secondary_use','has_secondary_use_agriculture','has_secondary_use_hotel','has_secondary_use_rental',\
'has_secondary_use_institution','has_secondary_use_school','has_secondary_use_industry','has_secondary_use_health_post',\
'has_secondary_use_gov_office','has_secondary_use_use_police','has_secondary_use_other'],axis=1)
```

```python
test = test.drop(labels=['building_id','geo_level_2_id','geo_level_3_id','count_floors_pre_eq',\
                        'has_superstructure_other','count_families','legal_ownership_status_C0',\
                        'legal_ownership_status_C1','legal_ownership_status_C2'],axis=1)
```

```python
test.info()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
from catboost import Pool, CatBoostClassifier
```

```python
train_labels=pd.read_csv('settrainlabels.csv')
```

```python
x = np.array(train)
y = np.array(train_labels)
```

```python
kf = KFold(n_splits=3)
for train_index, test_index in kf.split(x, y):
    model = CatBoostClassifier(iterations=100, random_state=42,loss_function='MultiClass').fit(
    x[train_index], y[train_index])
    score = model.score(x[test_index], y[test_index])
    print("accuracy=%.2f%%" % ( score * 100))
```

```python
## Se tiene un presicion del 48%
predict = model.predict(np.array(test))
```

```python
submit_format = pd.read_csv('submission_format.csv', index_col='building_id')
submission= pd.DataFrame(
    data=predict,
    columns=submit_format.columns,
    index=submit_format.index
)
```
