¿Qué región fue la más afectada? Agrupar por geo_level_1_id, geo_level_2_id y geo_level_3_id 

```python
import datetime as datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
df = pd.read_csv('dataset_v1')
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
df_regiones_level_1 = df.loc[:, ['geo_level_1_id', 'damage_grade']]
df_regiones_level_2 = df.loc[:, ['geo_level_2_id', 'damage_grade']]
df_regiones_level_3 = df.loc[:, ['geo_level_3_id', 'damage_grade']]
```

```python
total_damaged = df_regiones_level_1['damage_grade'].count()
```

## Nivel 1

```python
damage_count_level_1 = df_regiones_level_1.groupby('geo_level_1_id').count().rename(columns={'damage_grade': 'Damage count'})
damage_count_level_1.plot(kind='bar', figsize=(15, 5), xlabel = "Nivel geografico 1", title="Cantidad de edificios dañados por area (nivel 1)")
```

```python
damage_sum_level_1 = df_regiones_level_1.groupby('geo_level_1_id').sum().rename(columns={'damage_grade': 'Damage sum'})
damage_sum_level_1.plot(kind='bar', figsize=(15,5), xlabel="Nivel geografico 1", title="Nivel de daño total en nivel 1")
```

```python
damage_1_level_1 = df_regiones_level_1.loc[df_regiones_level_1['damage_grade'] == 1].groupby('geo_level_1_id').count().rename(columns={'damage_grade': 'Damage grade 1'})
damage_2_level_1 = df_regiones_level_1.loc[df_regiones_level_1['damage_grade'] == 2].groupby('geo_level_1_id').count().rename(columns={'damage_grade': 'Damage grade 2'})
damage_3_level_1 = df_regiones_level_1.loc[df_regiones_level_1['damage_grade'] == 3].groupby('geo_level_1_id').count().rename(columns={'damage_grade': 'Damage grade 3'})

damage_grades_level_1 = cant_regiones_level_1.join(damage_1_level_1, how='left').join(damage_2_level_1, how='left')\
                                             .join(damage_3_level_1, how='left').drop(columns=['damage_grade']).fillna(0)
damage_grades_level_1.plot(kind='bar', figsize=(20,10),
                           xlabel="Nivel geografico 1")
```

## Nivel 2

```python
damage_count_level_2 = df_regiones_level_2.groupby('geo_level_2_id').count().rename(columns={'damage_grade': 'Damage count'})
damage_count_level_2.info()
```

Tener en cuenta que este dataframe tiene 1414 columnas y el rango original de valores para 'geo_level_2_id'
es entre 0 y 1427 (rango de 1428 valores). Hay filas faltantes: implica que no hay datos registrados de ciertas
regiones o que directamente no hubo daños?

```python
damage_count_level_2.plot(kind='area', figsize=(15,5),
                           xlabel = "Nivel geografico 2", title="Cantidad de edificios dañados por area (nivel 2)")
```

Debido a la cantidad de datos, es dificil visualizar con este grafico la distribucion de edificios dañados
segun la region de nivel 2 en detalle. Por esta razón y para observar con más detalle cuáles fueron las regiones más afectadas
se decidió tomar un "top" de estos valores. Se tomaron los 30 valores más altos de 'Damaged count', que es un valor
en el orden de los datos de regiones de nivel 1

```python
damage_count_level_2_top_30 = damage_count_level_2.sort_values('Damage count', ascending=False).head(30).sort_index()
damage_count_level_2_top_30.plot(kind='bar', figsize=(15,5),
                                  xlabel="Nivel geografico 2", title="Top 30 de cantidad de edificios dañados por area (nivel 2)")
```

```python
damage_sum_level_2 = df_regiones_level_2.groupby('geo_level_2_id').sum().rename(columns={'damage_grade': 'Damage sum'})
damage_sum_level_2.plot(kind='area', figsize=(15,5), xlabel="Nivel geografico 2", title="Suma de daño total en nivel 2")
```

Como ocurre lo mismo que en el dataset anterior (con la agregación de conteo) hacemos el mismo procedimiento de
tomar los 30 valores mas altos.

```python
damage_sum_level_2_top_30 = damage_sum_level_2.sort_values('Damage sum', ascending=False).head(30).sort_index()
damage_sum_level_2_top_30.plot(kind='bar', figsize=(15,5), xlabel="Nivel geografico 2", title="Top 30 de suma de daño total en nivel 2")
```

Para obtener una visualizacion similar a los daños recibidos en la regiones de nivel 1,
donde se comparaba la cantidad de ocurrencias de cada valor de damage grade por region,
partiremos del dataframe con las 30 regiones de nivel 2 que recibieron el mayor total de daños
y calcularemos cuantas ocurrencias hubo de cada 'damage_grade' para este top 30

```python
damage_1_level_2 = df_regiones_level_2.loc[df_regiones_level_2['damage_grade'] == 1].groupby('geo_level_2_id').count().rename(columns={'damage_grade': 'Damage grade 1'})
damage_2_level_2 = df_regiones_level_2.loc[df_regiones_level_2['damage_grade'] == 2].groupby('geo_level_2_id').count().rename(columns={'damage_grade': 'Damage grade 2'})
damage_3_level_2 = df_regiones_level_2.loc[df_regiones_level_2['damage_grade'] == 3].groupby('geo_level_2_id').count().rename(columns={'damage_grade': 'Damage grade 3'})

damage_grades_level_2 = damage_sum_level_2_top_30.join(damage_1_level_2, how='left').join(damage_2_level_2, how='left')\
                                                 .join(damage_3_level_2, how='left').drop(columns=['Damage sum'])
damage_grades_level_2.plot(kind='bar', figsize=(20,10), xlabel="Nivel geografico 2")
```

```python
damage_sum_level_3 = df_regiones_level_3.groupby('geo_level_3_id').sum().rename(columns={'damage_grade': 'Damage sum'})
damage_sum_level_3.plot(kind='area', figsize=(15,5), xlabel="Nivel geografico 3", title="Suma de daño total en nivel 3")
```

Y el top 30

Por último, repetimos el gráfico comparando la cantidad de ocurrencias de cada valor de 'damage_grade'
para cada región (top 30)

```python
damage_1_level_3 = df_regiones_level_3.loc[df_regiones_level_2['damage_grade'] == 1].groupby('geo_level_3_id').count().rename(columns={'damage_grade': 'Damage grade 1'})
damage_2_level_3 = df_regiones_level_3.loc[df_regiones_level_2['damage_grade'] == 2].groupby('geo_level_3_id').count().rename(columns={'damage_grade': 'Damage grade 2'})
damage_3_level_3 = df_regiones_level_3.loc[df_regiones_level_2['damage_grade'] == 3].groupby('geo_level_3_id').count().rename(columns={'damage_grade': 'Damage grade 3'})

damage_grades_level_3 = damage_sum_level_3_top_30.join(damage_1_level_3, how='left').join(damage_2_level_3, how='left')\
                                                 .join(damage_3_level_3, how='left').drop(columns=['Damage sum'])
damage_grades_level_3.plot(kind='bar', figsize=(20,10), xlabel="Nivel geografico 3")
```
